```python
#Import all needed Packages

import matplotlib.pyplot as plt

import pandas as pd

import numpy as np
```

```python
#Set path to .csv file
IMDB_df = pd.read_csv('IMDB_Horror_Movies.csv')
IMDB_df.head()
```

```python
#Clean .csv file
IMDB_df.info()
```

```python
#changed dtype for Review Rating to str
IMDB_dfb = {'Review Rating': str}

IMDB_bfr = IMDB_bfr.astype(IMDB_dfb)

IMDB_bfr.dtypes
```

```python
#displaying the data
IMDB_bfr = IMDB_df[["Release Country", "Review Rating"]].dropna()
IMDB_bfr
```

```python
countries_IMDB = ["USA", "Japan", "Spain", "Germany", "India"]

IMDB_usa = IMDB_bfr[IMDB_bfr['Release Country'] == "USA"]["Review Rating"].mean()
IMDB_jpn = IMDB_bfr[IMDB_bfr['Release Country']== "Japan"]["Review Rating"].mean()
IMDB_spn = IMDB_bfr[IMDB_bfr['Release Country']== "Spain"]["Review Rating"].mean()
IMDB_gmn = IMDB_bfr[IMDB_bfr['Release Country']== "Germany"]["Review Rating"].mean()
IMDB_ind = IMDB_bfr[IMDB_bfr['Release Country']== "India"]["Review Rating"].mean()

relrat = [IMDB_usa, IMDB_jpn, IMDB_spn, IMDB_gmn, IMDB_ind]

#visualize this in a histogram
plt.style.use('seaborn-deep')
plt.hist(relrat, color = 'y', edgecolor = 'r')
plt.title("Country and Average Rating")
plt.xlabel("Country")
plt.ylabel("Review Rating")
plt.xticks([USA, JPN, SPN, GMN, IND], countries_IMDB)
plt.margins(x = 0.1, y = 0.1)
plt.show()
```

```python
ratings_test = IMDB_df.groupby(['Release Country'])['Review Rating'].count()
ratings_test.sort_values(ascending=False).plot.bar()
plt.xticks([USA, Japan, Spain, Germany, India], rotation=-90)
plt.show()
```

```python
#Total USA
usa = IMDB_df[(IMDB_df["Release Country"] == 'USA')]
usa_mn = usa['Review Rating'].mean()
usa_min = usa['Review Rating'].min()
usa_max = usa['Review Rating'].max()
usa_sum = usa['Review Rating'].sum()

#the number of American movies produced
usa_movies_total = IMDB_df['Release Country'].value_counts()['USA']


#Japanese
japan = IMDB_df[(IMDB_df["Release Country"] == 'Japan')]
jap_mn = japan['Review Rating'].mean()
jap_min = japan['Review Rating'].min()
jap_max = japan['Review Rating'].max()
jap_sum = japan['Review Rating'].sum()

# to get the number of japanese movies produced
jap_movies_total = IMDB_df['Release Country'].value_counts()['Japan']
                                                 
spain = IMDB_df[(IMDB_df["Release Country"] == 'Spain')]
span_mn = spain['Review Rating'].mean()
span_min = spain['Review Rating'].min()
span_max = spain['Review Rating'].max()
span_sum = spain['Review Rating'].sum()

#the number of Spanish movies produced
span_movies_total = IMDB_df['Release Country'].value_counts()['Spain']

#German
german = IMDB_df[(IMDB_df["Release Country"] == 'Germany')]
gmn_mn = german['Review Rating'].mean()
gmn_min = german['Review Rating'].min()
gmn_max = german['Review Rating'].max()
gmn_sum = german['Review Rating'].sum()

# to get the number of German movies produced
gmn_movies_total = IMDB_df['Release Country'].value_counts()['Germany']

#Indian
india = IMDB_df[(IMDB_df["Release Country"] == 'India')]
ind_mn = india['Review Rating'].mean()
ind_min = india['Review Rating'].min()
ind_max = india['Review Rating'].max()
ind_sum = india['Review Rating'].sum()

# to get the number of Indian movies produced
ind_movies_total = IMDB_df['Release Country'].value_counts()['India']
```

```python
x = ['USA', 'Japan', 'Spain', 'Germany', 'India']
y = [usa_movies_total, jap_movies_total, span_movies_total, gmn_movies_total, ind_movies_total]

plt.bar(x,y, color = '#196bde', width = 0.5)
plt.title('Top 5 Countries and Ratings')
plt.show()
```

```python
n = np.arange(2)
total = [usa_movies_total, jap_movies_total, span_movies_total, gmn_movies_total, ind_movies_total]
mean = [usa_mn, jap_mn, span_mn, gmn_mn, ind_mn]
width = 0.20
  
# plot data in grouped manner of bar type
plt.bar(n-0.2, total, width, color = '#196bde')
plt.bar(n+0.2, mean, width, color = '#61de19')
plt.title("Top Countries movies grouped with Average Rating")
plt.xticks(n, ["USA", "Japan", "Spain", "Germany", "Indian"])

plt.xlabel("Total and Average Review Rating ")
plt.ylabel("Scores")
plt.legend(["Total Movies", "Average Review Rating"])

plt.show()
```

```python
for x in range(1,9):
    IMDB_Clean2_df[f'Genre_{x}'] = [np.nan for y in range(len(IMDB_Clean2_df))]
IMDB_Clean2_df.head()
```

```python
IMDB_Clean3_df= IMDB_Clean2_df.reset_index(drop=True)
IMDB_Clean3_df.head()
```

```python
for x,y in enumerate(IMDB_Clean3_df['Genres'].tolist()):
    try:
        temp_var = y.split('|').replace(' ','')
        if len(temp_var) == 1:
            continue
        else:
            temp_var.remove(temp_var.index('Horror'))
            
        for i,j in enumerate(temp_var):

            IMDB_Clean3_df.at[x,f"Genre_{i+1}"]=j
    except:
        continue
```
