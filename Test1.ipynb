```python
#Import all needed Packages

import matplotlib.pyplot as plt

import pandas as pd

import numpy as np
```

```python
#Set path to .csv file
IMDB_df = pd.read_csv('IMDB_Horror_Movies.csv')
IMDB_df.head()
```

```python
#Clean .csv file
IMDB_df.info()
```

```python
IMDB_df.describe()
```

```python
IMDB_Clean1_df = IMDB_df[["Title", "Genres", "Release Date","Release Country" ,"Review Rating", "Filming Locations"]]
IMDB_Clean1_df.head()
```

```python
IMDB_Clean1_df.info()
```

```python
IMDB_Clean2_df = IMDB_Clean1_df.dropna()
IMDB_Clean2_df.info()
```

```python
IMDB_Clean2_df.Genres.value_counts()
```

```python
var = [len(x.split('|'))for x in IMDB_Clean2_df['Genres'].tolist()]
var.sort()
var[-10:]
```

```python
IMDB_Clean2_df["Release Date"].head(25)
```

```python
for x in range(1,9):
    IMDB_Clean2_df[f'Genre_{x}'] = [np.nan for y in range(len(IMDB_Clean2_df))]
IMDB_Clean2_df.head()
```

```python
IMDB_Clean3_df= IMDB_Clean2_df.reset_index(drop=True)
IMDB_Clean3_df.head()
```

```python
for x,y in enumerate(IMDB_Clean3_df['Genres'].tolist()):
    try:
        temp_var = y.split('|').replace(' ','')
        if len(temp_var) == 1:
            continue
        else:
            temp_var.remove(temp_var.index('Horror'))
            
        for i,j in enumerate(temp_var):

            IMDB_Clean3_df.at[x,f"Genre_{i+1}"]=j
    except:
        continue
```

```python
IMDB_Clean3_df.head()
```

```python
IMDB_Clean4_df = IMDB_Clean3_df[['Day', 'Month', 'Year']] =  IMDB_Clean3_df.str.split("-", expand = True)
print(IMDB_Clean4_df.count())
IMDB_Clean4_df.head()
```

```python
IMDB_Clean4_df['Release Year'] = pd.DatetimeIndex(IMDB_Clean4_df['Release Date']).year
IMDB_Clean4_df.head()
```
